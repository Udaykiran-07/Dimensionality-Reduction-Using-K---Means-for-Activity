```python
import requests
from bs4 import BeautifulSoup
import zipfile
import io
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import time

# Function to download and load dataset
def load_data():
    page_url = 'https://archive.ics.uci.edu/dataset/240/human+activity+recognition+using+smartphones'
    page_response = requests.get(page_url)
    if page_response.status_code == 200:
        soup = BeautifulSoup(page_response.content, 'html.parser')
        download_link = soup.select_one('a[href$=".zip"]')['href']
        full_download_url = 'https://archive.ics.uci.edu' + download_link
        response = requests.get(full_download_url)
        if response.status_code == 200:
            with zipfile.ZipFile(io.BytesIO(response.content)) as outer_zip:
                inner_zip_name = 'UCI HAR Dataset.zip'
                with outer_zip.open(inner_zip_name) as inner_zip_file:
                    with zipfile.ZipFile(io.BytesIO(inner_zip_file.read())) as inner_zip:
                        with inner_zip.open('UCI HAR Dataset/train/X_train.txt') as myfile:
                            df = pd.read_csv(myfile, delim_whitespace=True, header=None)
                        with inner_zip.open('UCI HAR Dataset/train/y_train.txt') as myfile_y:
                            y = pd.read_csv(myfile_y, delim_whitespace=True, header=None)
    else:
        raise Exception("Failed to download or parse the dataset.")
    return df, y

# Call the load_data function to load the dataset
df, y = load_data()

# Print dataset shapes
print(f"Feature dataset shape: {df.shape}")
print(f"Target dataset shape: {y.shape}")

# Display first few rows of the features
print("\nFirst few rows of the feature dataset:")
print(df.head())

# Display first few rows of the target
print("\nFirst few rows of the target dataset:")
print(y.head())

# Summary of the feature dataset
print("\nFeature dataset summary:")
print(df.describe())

# Check for missing values
print("\nMissing values in the dataset:")
print("Features:", df.isnull().sum().sum())
print("Target:", y.isnull().sum().sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Compute variance for each feature
variances = df.var()

# Plot variances
plt.figure(figsize=(10, 6))
plt.plot(variances, marker='o', linestyle='none')
plt.title("Feature Variances")
plt.xlabel("Feature Index")
plt.ylabel("Variance")
plt.show()

# Identify features with low variance
low_variance_features = variances[variances < 0.01]
print(f"\nNumber of low variance features: {len(low_variance_features)}")
print("Indices of low variance features:", low_variance_features.index.tolist())

# Correlation matrix
correlation_matrix = df.corr()

# Plot heatmap of correlations
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', vmax=1.0, vmin=-1.0)
plt.title("Feature Correlation Matrix")
plt.show()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Create a LabelEncoder object
encoder = LabelEncoder()

# Fit the encoder to your target variable and transform it
encoded_y = encoder.fit_transform(y)

# Split the data into training and testing sets (80% training, 20% testing)
X_train_full, X_test_full, y_train, y_test = train_test_split(
    df_scaled, encoded_y, test_size=0.2, random_state=42
)

# Check the shape of the resulting datasets
print(f"Training set shape: {X_train_full.shape}")
print(f"Test set shape: {X_test_full.shape}")
```
